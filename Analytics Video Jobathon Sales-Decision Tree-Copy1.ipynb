**Data Cleaning**

```python
x = train.drop(['#Order','Sales'],axis =1)
y = train[['#Order','Sales']]
```

```python
x.head()
```

```python
y.head()
```

**Model Building**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(x)
```

```python
#Scaling the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
import xgboost
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error
```

```python
LR = LinearRegression()
DT = DecisionTreeRegressor()
RF = RandomForestRegressor()
XGB = XGBRegressor()
ABR = AdaBoostRegressor()
```

**Model Building**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.20, random_state = 500)
```

```python
#training with the all of them

LR.fit(x_train,y_train)
RF.fit(x_train,y_train)
DT.fit(x_train,y_train)

print("All Models Trained")
```

```python
print('Logistic regression accuracy =', LR.score(x_test,y_test))
print('Logistic regression MSE =', mean_squared_error(y_test,LR.predict(x_test)))
print('Logistic regression MSLE =', mean_squared_log_error(y_test,LR.predict(x_test)))
print('Logistic regression MAE =', mean_absolute_error(y_test,LR.predict(x_test)))

print('Random Forest accuracy =',RF.score(x_test,y_test))
print('Random Forest MSE =', mean_squared_error(y_test,RF.predict(x_test)))
print('Random Forest MSLE =', mean_squared_log_error(y_test,RF.predict(x_test)))
print('Random Forest MAE =', mean_absolute_error(y_test,RF.predict(x_test)))

print('Decision Tree accuracy=',DT.score(x_test,y_test))
print('Decision Tree MSE=', mean_squared_error(y_test,DT.predict(x_test)))
print('Decision Tree MSLE=', mean_squared_log_error(y_test,DT.predict(x_test)))
print('Decision Tree MAE=', mean_absolute_error(y_test,DT.predict(x_test)))
```

**HyperParameter Tuning**

```python
from sklearn.model_selection import GridSearchCV
```

```python
RF_Params = {'min_samples_leaf': list(range(2,6)),
         'min_samples_split': list(range(2,6)),
         'max_depth':list(range(2,6)),
            'random_state':[900]} 
```

```python
DT_Params = {'splitter': ['best','random'],
          'min_samples_leaf': list(range(2,12)),
         'min_samples_split': list(range(2,12)),
         'max_depth':list(range(2,12)),
            'random_state':[1100,200]}
```

```python
RF_grid = GridSearchCV(RF, param_grid= RF_Params, cv = 2, n_jobs=-1)
```

```python
RF_grid.fit(x_train,y_train)
print('Best Estimators == > \n', RF_grid.best_params_)
RF_clf= RF_grid.best_estimator_
print("Decision Tree Regression",RF_clf.fit(x_train, y_train))
```

```python
print('Random Forest Accuracy=', RF_clf.score(x_test,y_test))
print('Random Forest MSE =', mean_squared_error(y_test,RF_clf.predict(x_test)))
print('Random Forest MSLE =', mean_squared_log_error(y_test,RF_clf.predict(x_test)))
print('Random Forest MAE=', mean_absolute_error(y_test,RF_clf.predict(x_test)))
```
